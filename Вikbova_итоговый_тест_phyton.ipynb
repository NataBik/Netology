## Задания 5 и 6

- Показать распределение сотрудников по зарплатам
- Показать распределение сотрудников по зарплатам в каждом
департаменте по отдельности

```python
# сводная таблица с функцией итогов позволяет ответить сразу на оба вопроса

df_by_salary = df.pivot_table(index = 'salary', columns = 'department', values = 'satisfaction_level', aggfunc = 'count', margins = True).head()
df_by_salary
```

Преобразуем таблицу в формат процентных соотношений количества сотрудников по категорям зарплат в каждом департаменте и итого

```python
list_departments = list(df_by_salary.columns)
list_departments
```

```python
list_high = df_by_salary.loc['high']
list_medium = df_by_salary.loc['medium']
list_low = df_by_salary.loc['low']
list_ALL = df_by_salary.loc['All']
list_high
```

```python
# расчет процентов
number_departments = len(list_high)
clicker = [i for i in range(0, number_departments)]

list_high_percent = []
list_medium_percent = []
list_low_percent = []

for n in clicker:
  list_high_percent.append(round(list_high[n]/list_ALL[n]*100,1))
  list_medium_percent.append(round(list_medium[n]/list_ALL[n]*100,1))
  list_low_percent.append(round(list_low[n]/list_ALL[n]*100,1))
```

```python
data = {'Department': list_departments, 'High_salary_%': list_high_percent, 'Medium_salary_%': list_medium_percent, 'Low_salary_%':list_low_percent}

df_by_salary_new = pd.DataFrame(data)
df_by_salary_new['Total_%'] = round(df_by_salary_new['High_salary_%'] + df_by_salary_new['Medium_salary_%'] + df_by_salary_new['Low_salary_%'],0)
df_by_salary_new
```

## Задание 7

Проверить гипотезу, что сотрудники с высоким окладом
проводят на работе больше времени, чем сотрудники с низким
окладом

```python
# определим среднее время, которое проводят на работе, разные группы сотрудников

df.pivot_table(index = 'salary', columns = 'department', values = 'average_montly_hours', aggfunc = 'mean', margins = True).head()
```

Как видно из таблицы выше - в разных департаментах ситуация разная, однозначной зависимости между уровнем зарплаты и временем, проводимым на работе, вероятно нет.

В целом по компании, скорее наоборот, сотрудники с высоким окладом проводят в среднем меньше времени на работе, чем сотрудники с низким окладом.

Чтобы определить, является ли разница достоверной (значимой), проведем стат.тестирование.

```python
df_high = df[df['salary']=='high'] # только сотрудники со статусом зарплаты 'high'
df_low = df[df['salary']=='low'] # только сотрудники со статусом зарплаты 'high'
```

```python
df_high['average_montly_hours'].describe()
```

```python
df_low['average_montly_hours'].describe()
```

```python
# По умолчанию критерий Стьюдента подразумевает равенство дисперсий двух выборок. Если дисперсии отличаются (для группы "high" std = 47,71, а для группы "low" std = 50,8), то нужно использовать поправку Уэлча (Welch) 'ttest_ind'.

# Нулевая гипотеза: средние значения в двух группах сотрудников с разным уровнем зарпалты достоверно не отличаются

import scipy.stats as st
stat, p = st.ttest_ind(df_high['average_montly_hours'], df_low['average_montly_hours'], equal_var = False)

print(f"Статистика = {stat:.3f}, p = {p:.3f}")
if p > 0.05:
    print('Не отклоняем нулевую гипотезу, средние, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, средние, вероятно, различаются')
```

## Задание 8

Рассчитать следующие показатели среди уволившихся и не
уволившихся сотрудников (по отдельности):

* Доля сотрудников с повышением за последние 5 лет
* Средняя степень удовлетворенности
* Среднее количество проектов

```python
# Определим долю сотрудников с повышением за последние 5 лет среди работающих сотрудников

df_left_0 = df[df['left']==0] # только сотрудники со статусом 'left' = 0 (не уволившиеся)

import numpy as np
df_left_0['Promotion_status'] = np.where(df_left_0['promotion_last_5years']==0, 'not promoted', 'promoted') # добавлен столбец со статусом наличия/отсутствия повышения для последующей фильтрации по нему

df_left_0_filtered = df_left_0[['promotion_last_5years', 'Promotion_status']].groupby(['Promotion_status']).count() #таблица с разбивкой по сотрудникам с повышением и без
df_left_0_filtered['Total'] = df_left_0_filtered['promotion_last_5years'].sum() # вычисляемый столбец с общим количеством сотрудников
df_left_0_filtered['Share-promoted'] = round((df_left_0_filtered['promotion_last_5years']/df_left_0_filtered['Total'])*100, 0) # вычисляемый столбец с оценкой доли сотрудников с повышением и без

df_left_0_filtered.head()
```

**3% работающих сотрудников получили повышение за последние 5 лет**

```python
# Аналогично определим долю сотрудников с повышением за последние 5 лет среди уволившихся сотрудников

df_left_1 = df[df['left']==1] # только сотрудники со статусом 'left' = 1 (уволившиеся)

import numpy as np
df_left_1['Promotion_status'] = np.where(df_left_1['promotion_last_5years']==0, 'not promoted', 'promoted')

df_left_1_filtered = df_left_1[['promotion_last_5years', 'Promotion_status']].groupby(['Promotion_status']).count()
df_left_1_filtered['Total'] = df_left_1_filtered['promotion_last_5years'].sum()
df_left_1_filtered['Share-promoted'] = round((df_left_1_filtered['promotion_last_5years']/df_left_1_filtered['Total'])*100, 0)

df_left_1_filtered.head()
```

**Среди уволившихся сотрудников процент повышения составлял 1%**

```python
# Рассчитаем средние показатели удовлетворенности и нагрузки проектами среду уволившихся и работающих сотрудников

df[['satisfaction_level', 'number_project', 'left']].groupby(['left']).mean().head()
```

**Среди уволившихся сотрудников степень удовлетворенности заметно ниже (0.4), чем среди работающих (0.7)**

**Среднее количество проектов среди уволившихся сотрудников несколько выше (3,9), чем среди работающих (3,8)**

## Задание 9

Разделить данные на тестовую и обучающую выборки
Построить модель LDA, предсказывающую уволился ли
сотрудник на основе имеющихся факторов (кроме department и
salary)
Оценить качество модели на тестовой выборке

```python
import sklearn
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
y = df['left']
X = df[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years']]
```

```python
# разделение на тестовую и обучающую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# масштабирование

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
 # Выполнение LDA

lda = LDA(n_components=1)
X_train = lda.fit_transform(X_train, y_train)
X_test = lda.transform(X_test)
```

```python
# предсказание на тестовых данных
classifier = RandomForestClassifier(max_depth=2, random_state=0)

classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
# оценка качества модели
print('Точность модели: ' + str(accuracy_score(y_test, y_pred)))
```

Точность предсказания модели 81% (приемлемая)
